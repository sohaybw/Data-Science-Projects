```python
# Visualizing the Frequency of (sex ) Column
```

```python
explode = (0, 0.1)  
fig1, ax1 = plt.subplots()
ax1.pie(data.sex.value_counts().values, explode=explode, 
        labels=data.sex.value_counts().index, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  
plt.tight_layout()
plt.show()
# Visualizing the Frequency of gender due to 'sex' column
```

```python
plt.figure(figsize=(15,5))
plt.title('Infection reason')
data[data.infection_reason != 'nan'].infection_reason.value_counts().plot.bar();
# Visualizing the Various Reasons of infection to Detect the Most Effective Reason 
```

```python
# Visualizing  a figure to show the state after infection comparing to the age and gender
fig_dims = (10, 10)
fig, ax = plt.subplots(figsize=fig_dims)
sns.barplot(x='state',y='Age',data=data,hue='sex',palette='Paired')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
```
